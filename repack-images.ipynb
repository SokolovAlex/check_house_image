```python
import os
import shutil

fromFolder = "./open_images_classification"
dist = "./open_images_house_1"
lerningCategories = ["House", "Building", "Cabinetry", "Furniture", "Desk", "Door", "Kitchen appliance", "Window", "Office supplies", "Cupboard"]
otherFolder = "Other"

# fromFolder = "./open_images_house"
# dist = "./open_images_house2"
# lerningCategories = ["Other"]
# otherFolder = "House"

for root, subfolders, files in os.walk(fromFolder, followlinks=False):
    if len(subfolders) == 0:
        category = root.split("\\")[-1]
        print("- category ->", category)
        print("- files ->", files)
        for item in files:
            if item.endswith(".jpg") :
                print("item  -> ", item)
                categoryName= category if category in lerningCategories else otherFolder
                fileNamePath = os.path.join(root, item)
                distPath = os.path.join(dist, categoryName, f"{category}_{item}")
                distDirectory = os.path.join(dist, categoryName)
                print("categoryName  -> ", categoryName)
                print("distPath  -> ", distPath)
                
                if not os.path.exists(distDirectory):
                    os.makedirs(distDirectory)
                
                shutil.copy(fileNamePath, distPath)
```
